```python
from math import ceil

import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.layouts import gridplot
from bokeh.palettes import all_palettes, Inferno, YlGnBu
```

```python
n = 500
x = np.random.standard_normal(n)
y = np.random.standard_normal(n)
```

```python
palette = Inferno[256]
p = figure(background_fill_color=palette[1], width=1024, height=1024)
p.grid.visible = False
p.hexbin(x, y, size=0.5, palette=palette)
p.circle(x, y, color="white", size=1)
output_notebook()
show(p)
```

```python
def display_all_palettes(palette_dict=all_palettes):
    p = []
    for name, palette in palette_dict.items():
        print(name, palette)
        if isinstance(palette, dict):
            color = palette[max(palette.keys())]
        else:
            color = palette
        p.append(figure(width=500, height=50, toolbar_location=None, title=name))
        p[-1].rect(x=list(range(len(color))), y=[0]*len(color), width=1, height=1, color=color)
        p[-1].axis.visible = False
        p[-1].grid.visible = False
        p[-1].outline_line_color = None
        p[-1].x_range.start = -0.5
        p[-1].x_range.end = len(color) - 0.5
        p[-1].y_range.start = -0.5
        p[-1].y_range.end = 0.5
        
    # Arrange plots in a grid
    grid = gridplot(p, ncols=1, width=500, height=50)
    output_notebook()
    show(grid)
```

```python
display_all_palettes()
```
